# Data Cleaning & Preparation

In this section, we prepare the dataset for analysis and modeling by:
- Handling missing values
- Fixing data types
- Removing unnecessary columns
- Creating clean and consistent features

## Load Dataset

We begin by loading the dataset into a pandas DataFrame to explore its structure and contents.

```python
import pandas as pd
import numpy as np

df=pd.read_excel(r'Telco_customer_churn.xlsx')
```

## Removing Unnecessary Columns

Some columns are redundant or not useful for analysis.

We remove:
- Duplicate target columns
- Columns that do not add predictive value

This simplifies the dataset and avoids confusion during modeling.

```python
df["Churn"] = df["Churn Value"]
df.drop(["Churn Label", "Churn Value"], axis=1, inplace=True)
```

```python
columns_to_drop = [
    "Gender",
    "CustomerID",
    "Count",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Lat Long",
    "Latitude",
    "Longitude",
    "Churn Score",
    "CLTV",
    "Churn Reason"
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")
```

## Handling Missing Values

Missing values can negatively impact analysis and model performance.

We identify and handle missing data to ensure:
- Data consistency
- Reliable results

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

Missing values were removed for simplicity. In a real-world scenario, imputation techniques could be used to preserve more data.
The dataset is now free of missing values and ready for further processing.


## Duplicate Handling

After removing the CustomerID column, 43 duplicate rows were identified.

Since CustomerID was the unique identifier, removing it caused some rows to appear identical.

These duplicates were removed to avoid bias in the model.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Encoding Categorical Variables

Machine learning models require numerical input.

We convert categorical values (e.g., Yes/No) into numeric format to make them usable for modeling.
Binary variables are converted to 1 (Yes) and 0 (No) for easier interpretation and modeling.

```python
yes_no_cols = df.select_dtypes(include="object").columns

for col in yes_no_cols:
    if df[col].nunique() == 2:
        df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
df.replace("No internet service", "No", inplace=True)
df.replace("No phone service", "No", inplace=True)
```

```python
dummies = pd.get_dummies(df[[
    'Contract',
    'Internet Service',
    'Payment Method'
]])

df = pd.concat([df, dummies], axis=1)
```

```python
df.info()
df.shape
df.describe()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

At this stage:
- The dataset is clean and consistent
- Missing values have been handled
- Categorical variables are encoded
